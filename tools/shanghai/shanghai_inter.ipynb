```python
import pandas as pd
import os
import glob
import xarray as xr
from datetime import datetime, timedelta

initial_time = "2024-06-16T12:00:00"
data_root = '../../data/output'

data = pd.read_csv("20240616160000_20240624000000.csv",encoding='gbk')  # 替换为您的CSV文件路径
# 提取唯一的站点ID
station_ids = data['区站号(字符)'].unique()
filtered_list = [s for s in station_ids if s.isdigit()]
print(f'the number of station_id  is {len(filtered_list)}')
target_datetime = pd.date_range(start="2024-06-16T12", end="2024-06-23T12", freq='6h')
# print(target_datetime)

for station_id in station_ids[0:100]:
    station_data = data[data['区站号(字符)'] == station_id]
   
    station_data['日期'] = station_data.apply(lambda row: f"{row['年']}-{row['月']:02d}-{row['日']:02d}T{row['时']:02d}", axis=1)

    # print(station_data)
    
    # 获取文件夹下的所有文件
    file_lists = glob.glob(os.path.join(f'{data_root}/{initial_time}', "*surface.nc"))
    file_lists.sort()
    print(file_lists)
    lat = station_data['纬度'].iloc[0]
    lon = station_data['经度'].iloc[0]
    print(lat, lon)
    
    mae_list = []
    for file_name in file_lists:
        # pred_time = file_name.split('_')
        nc_data = xr.open_dataset(file_name)
        pred_time_utc = str(nc_data['time'].data[0]).split(':')[0]
        t2m_pred = nc_data['t2m'].interp(longitude=lon, latitude=lat).values-273.15
        

        date_time = datetime.strptime(pred_time_utc, "%Y-%m-%dT%H")
        real_time_bj = date_time + timedelta(hours=0)
        real_time_bj = real_time_bj.strftime("%Y-%m-%dT%H")
        
        t2m_gt = station_data[station_data['日期']==real_time_bj]['温度/气温'].values
        mae = t2m_pred-t2m_gt
        mae_list.append()
        print(pred_time_utc, t2m_pred, real_time_bj, t2m_gt)
        
     



    
    
```
